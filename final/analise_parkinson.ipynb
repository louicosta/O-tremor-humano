```python
translate = {
    'Entity': 'PaÃ­s',
    'Code_x': 'Sigla',
    'Year_x': 'Ano morte',
    'Death rate per 100,000 people - Cause: Parkinson disease - Sex: Both sexes - Age_group: ALLAges	': 'Taxa de mortalidade por 100 mil',
    'Year_y': 'Ano doenÃ§a',
    "Current number of cases of parkinson's disease per 100,000 people, in both sexes aged age-standardized": 'NÃºmero atual de casos por 100 mil',
}

new_columns = []
for chave, valor in translate.items():
    new_columns.append(valor)

print(new_columns)
```

```python
# Filtrando os dados para que mostre apenas o paÃ­s Brasil e anos a partir de 2012

df_filtered = df_final[(df_final['Ano morte'] > 2011) & (df_final['PaÃ­s'] == 'Brazil')]
df_filtered.shape
```

```python
sns.scatterplot(x='Ano morte', y='Taxa de mortalidade por 100 mil', data=df_filtered, hue='Taxa de mortalidade por 100 mil', palette='viridis')
plt.title('MÃ©dia de morte por ano no Brasil')
plt.xlabel('Ano morte')
plt.ylabel('Taxa de mortalidade por 100 mil')
plt.show()
```

```python
sns.scatterplot(x='Ano doenÃ§a', y='NÃºmero atual de casos por 100 mil', data=df_filtered, hue='NÃºmero atual de casos por 100 mil', palette='viridis')
plt.title('MÃ©dia de doenÃ§a por ano no Brasil')
plt.xlabel('Ano doenÃ§a')
plt.ylabel('NÃºmero atual de casos por 100 mil')
plt.show()
```
