```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_final = pd.read_csv('base_unificada_parkinson.csv')
df_final
```

```python
df_final.info()
```

```python
columns_drop = ['Code_y']
df_final = df_final.drop(columns_drop, axis=1)
df_final
```

```python
translate = {
    'Entity_x': 'PaÃ­s',
    'Code_x': 'Sigla',
    'Year': 'Ano',
    'Death rate per 100,000 people - Cause: Parkinson disease - Sex: Both sexes - Age_group: ALLAges	': 'Morte por 100 mil',
    'Entity_y': 'RegiÃ£o',
    "Current number of cases of parkinson's disease per 100,000 people, in both sexes aged age-standardized": 'DoenÃ§a por 100 mil',
}

new_columns = []
for chave, valor in translate.items():
    new_columns.append(valor)

print(new_columns)
```

```python
df_final.columns = new_columns
df_final
```

```python
# Supondo que vocÃª quer contar a frequÃªncia das combinaÃ§Ãµes Ãºnicas de 'Morte por 100 mil' e 'Ano'
count_type = df_final.groupby(['Morte por 100 mil', 'Ano']).size()
count_type.plot(kind='bar')
plt.title('MÃ©dia de morte por ano')
plt.xlabel('Morte por 100 mil')
plt.ylabel('Ano')
plt.show()
```

```python
count_type = df_final['Morte por 100 mil', 'Ano'].value_counts()
count_type.plot(kind='bar')
plt.title('Valor gasto por mÃ©todo de pagamento')
plt.xlabel('MÃ©todo de pagamento')
plt.ylabel('Valor')
```
