```python
import pandas as pd
```

## Coleta de dados

A base de dados tem várias tabelas diferentes relacionados a mortalidade, vivência de pessoas com doença de parkinson com anos dos estudos e localidade.

Neste notebook será realizado apena a junção das batabases iniciais e união para o dataset final.

```python
df_death = pd.read_csv('death-rate-from-parkinsons-disease-ghe.csv')
df_death
```

```python
df_disease = pd.read_csv('parkinsons-disease-prevalence.csv')
df_disease
```

## Unindo o dataset
Para a análise proposta, irei utilizar a coluna 'Year''

```python
df_unified = pd.merge(df_death,df_disease, on='Entity')
df_unified
```

## Esportando a base final unificada

```python
df_unified.to_csv('base_unificada_parkinson.csv', index=False)
```
